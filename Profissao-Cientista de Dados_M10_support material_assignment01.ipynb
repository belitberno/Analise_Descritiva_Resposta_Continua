Para isto, conduziu um estudo com alguns clientes, comprovando suas rendas através de olerites e outros documentos, e pretende construir um modelo preditivo para esta renda com base em algumas variáveis que já possui em seu banco de dados.

As variáveis são intuitivas - note que há uma variável 'index' que é um identificador do cliente e que em geral o ```read_csv``` do pandas coloca também uma variável sequencial.

Estes dados estão no arquivo ```previsao_de_renda.csv```. Carregue-o em um *dataframe*.

1) Avalie a estrutura de correlação das variáveis quantitativas através de um gráfico de "matriz de dispersão" e através da avaliação gráfica da matriz de correlações. Veja se você identifica algum padrão interessante ou que te faça sentido.

````python
base = pd.read_csv('previsao_de_renda.csv')
base.drop(['index', 'mau'],axis = 1, inplace = True)
base.head()
````

````python
base.corr()
````

````python
sns.pairplot(base.corr())
````

````python
cmap = sns.diverging_palette(h_neg=125, h_pos=350, as_cmap=True, sep = 60, center = 'light')

sns.clustermap(base.corr(), figsize=(10, 10), center = 0, cmap=cmap)
````

-> A quantidade de filhos e de pessoas na residencia são correlacionadas;
-> As variáveis idade, tempo de emprego e renda também são correlacionadas.

2) Avalie um gráfico de dispersão (*scatterplot*) específico para as duas variáveis mais correlacionadas com *renda*.

````python
renda_x_emprego=base[['renda','tempo_emprego']].copy()
sns.scatterplot(x='renda', y='tempo_emprego', data =renda_x_emprego)
````

````python
renda_x_idade=base[['renda','idade']].copy()
sns.scatterplot(x='renda', y='idade', data =renda_x_idade)
````

````python
renda_x_residencia=base[['renda','qt_pessoas_residencia']].copy()
sns.scatterplot(x='renda', y='qt_pessoas_residencia', data =renda_x_residencia)
````

3) Na su aopinião, há outliers na variável renda?

Sim.

4) Calcule o logaritmo da variável renda e repita a sequência de análise

````python
renda_temp_log = renda_x_emprego.copy()
renda_temp_log['log_renda'] = np.log(renda_temp_log['renda']) 

sns.scatterplot(x='log_renda', y='tempo_emprego', data = renda_temp_log)
````

````python
renda_idade_log = renda_x_idade.copy()
renda_idade_log['log_renda'] = np.log(renda_idade_log['renda']) 

sns.scatterplot(x='log_renda', y='idade', data = renda_idade_log)
````

````python
renda_residencia_log = renda_x_residencia.copy()
renda_residencia_log['log_renda'] = np.log(renda_residencia_log['renda']) 

sns.scatterplot(x='log_renda', y='qt_pessoas_residencia', data = renda_residencia_log)
````

5) A aplicação do LOG você acha que melhorou ou piorou a análise?

Sim, incluindo a variável de quantidade de pessoas na residencia, e com a aplicação do LOG ainda vemos claros exemplos de outliers, não resolveu por completo, mas trouxe uma melhora significativa para todas as variáveis
