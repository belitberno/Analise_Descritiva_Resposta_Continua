````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
base = pd.read_csv('previsao_de_renda.csv')
base.head()
````

````python
ax = sns.pointplot(x="posse_de_imovel", y="renda",
                   data=base, dodge=True, ci = 90)
````

````python
ax = sns.pointplot(x="posse_de_veiculo", y="renda",
                   data=base, dodge=True, ci = 90)
````

As duas variáveis apresentam uma diferença, bem expressiva, de renda para quem tem imovel e veiculos.
